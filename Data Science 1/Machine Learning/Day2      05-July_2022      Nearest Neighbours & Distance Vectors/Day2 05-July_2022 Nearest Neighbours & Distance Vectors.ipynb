```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Before we build any machine learning model we should do both machine training and machine testing.
# Data should be split into 70:30, 60:40, 80:20 ratio but train ratio should be higher than test ratio.
```

```python
df.head(3)
```

```python
df.shape
```

```python
x= df[['mass', 'width', 'height']] # independent variable
y= df['fruit_label']  # dependent variable
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.25, random_state=0)
# random_state is helps to randomize data, it will pick data randomly.
```

```python
x.shape  # it has 100% of data
```

```python
x_train.shape  # it has 75% of data
```

```python
x_test.shape  it has 25% of data
```

```python
y.shape # it has 100% of data
```

```python
y_train.shape # it has 75% of data
```

```python
y_test.shape # it has 25% of data
```

```python
knn = KNeighborsClassifier()  # here we saved 'KNeighbors' model into 'knn' variable
knn.fit(x_train, y_train)  # here we fit our data into 'KNeighbors' model 
```

```python
knn.score
```

```python
knn.score(x_test, y_test)  # here we tested our data into 'KNeighbors' model
```

```python
# above output has 0.6 output, it means 60% accuracy
```

```python
prediction = knn.predict([[6.8,5.7,8.7]]) # here we are predicting data, [6.8,5.7,8.7] is ['mass', 'width', 'height']
prediction
```

```python
predict[prediction[0]] # model is not working well
```

```python
prediction1 = knn.predict([[7.8, 5.7, 3.9]])  # model is not working well
predict[prediction1[0]]
```

```python
# Eucledian Distance ( )

from math import sqrt
r1 = [10,20, 15, 10,5]   # here we created two data and saved it in variable r1,r2
r2 = [12,24, 18, 8,7]

def ed(a,b):    # here we created a function
    return sqrt(sum((i-j)**2 for i,j in zip(a,b)))
```

```python
ed(r1,r2)  # here we passed above variable
```
