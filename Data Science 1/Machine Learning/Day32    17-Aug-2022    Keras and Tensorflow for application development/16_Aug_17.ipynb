```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
mnist_fashion = tf.keras.datasets.fashion_mnist
(x_train,y_train), (x_test, y_test) = mnist_fashion.load_data()
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size = 0.25, random_state = 0)
```

```python
from tensorflow.keras.layers import Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Flatten())
```

```python
model.add(Dense(512, activation= 'relu'))
model.add(Dense(256, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))
```

```python
from tensorflow.keras.optimizers import Adam, SGD, GD
adam = Adam()
loss = ['mean_squared_error', 'mean_absolute_error', 'sparse_categorical_crossentropy']
metrics = ['accuracy', 'precision', 'recall']
```

```python
model.compile(optimizer=adam, loss = loss[2], metrics = metrics[0])
```

```python
history = model.fit(x = x_train, y = y_train, batch_size = 128,epochs = 10)
```

```python
def plot_history(history):
  plt.plot(history.history['accuracy'])
  plt.title('Model Accuracy')
  plt.xlabel('Epochs')
  #plt.plot(history.history['val_accuracy'])
  plt.show()

  plt.plot(history.history['loss'])
  plt.title('Model Loss')
  plt.xlabel('Epochs')
  #plt.plot(history.history['val_loss'])
  plt.show()
```

```python
plot_history(history)
```
