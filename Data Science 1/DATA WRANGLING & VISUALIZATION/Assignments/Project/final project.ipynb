```python
3. Low Level Data Understanding:
    a. Find count of unique values in location column.
    b. Find which continent has maximum frequency using values
counts.
    c. Find maximum and mean value in 'total_cases'.
    d. Find 25%, 50% and 75% quartile value in 'total_deaths'.
    e. Find which continent has maximum
    'human_development_index'.
    f. Find which continent has minimum 'gdp_per_capita'.    
```

```python
# a. Find count of unique values in location column.

print("Number of unique values in location column is :",df['location'].nunique())
print(" ")
print(" ")
print("-----Count of unique values in location column is-----")
df['location'].value_counts()
```

```python
# b. Find which continent has maximum frequency using values

print("Maximum Frequensy is in a continent is : ", df['continent'].value_counts().max())

print(" ")
print("---**---All continents frquency using values---**---\n",df['continent'].value_counts())
```

```python
# c. Find maximum and mean value in 'total_cases'.

print("maximum value in 'total_cases' : ",df['total_cases'].max())
print(" ")
print("mean value in 'total_cases' : ",df['total_cases'].mean())
```

```python
# d. Find 25%, 50% and 75% quartile value in 'total_deaths'.

print("25% quartile value in 'total_deaths' : ",df['total_deaths'].quantile(.25))
print("50% quartile value in 'total_deaths' : ",df['total_deaths'].quantile(.50))
print("75% quartile value in 'total_deaths' : ",df['total_deaths'].quantile(.75))
```

```python
# e. Find which continent has maximum  
#    'human_development_index'.
print("Please find below report to know which continent has maximum number of 'human_development_index': ")
print(df.groupby('continent').count()['human_development_index'])



print(" ")
print(" ")
print("Please find below report to know which continent has maximum 'human_development_index'")
df.groupby('continent')['human_development_index'].max()
```

```python
# f. Find which continent has minimum 'gdp_per_capita'.

print("Please find below report to know which continent has minimum number of 'gdp_per_capita': ")
print(df.groupby('continent').count()['gdp_per_capita'])



print(" ")
print(" ")
print("Please find below report to know which continent has minimum 'gdp_per_capita' : ")
df.groupby('continent')['gdp_per_capita'].min()
```

```python
# 4. Filter the dataframe with only this columns
# ['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] 
# and update the data frame.

df=df.filter(items=['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'])
df.head()
```

```python
# 5. Data Cleaning
#   a. Remove all duplicates observations
#   b. Find missing values in all columns
#   c. Remove all observations where continent column value is 
# missing
#   Tip : using subset parameter in dropna
#   d. Fill all missing values with 0
```

```python
#   a. Remove all duplicates observations
df=df.drop_duplicates()
df
```

```python
#   b. Find missing values in all columns
print(" Please find below data, missing values will be shown as 'False' ")
print(" ")
print(df.isnull())
print(" ")
print(" ")

print(" Please find below data to know number of missing values each coulumns has ")
print(" ")
print(df.isnull().sum())
```

```python
#   c. Remove all observations where continent column value is 
# missing

df['continent'].dropna(inplace=True)
df['continent']
```
