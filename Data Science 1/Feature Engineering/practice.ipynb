## a. For Science assessment, analyze the data and share 5 most important insights

```python
df_Science =df[df['Subject']=='Science']
df_Science
```

```python

plt.figure(figsize=(8,5))
sns.countplot(x='Grade Level',hue='Gender',data=df_Science,palette='coolwarm')

print("Here we get know that huge number of students is in 'Not at Grade Level' and Female students is slightly better than Male  students")
```

```python
plt.figure(figsize=(10,5))
sns.set_context("talk",font_scale=1)
sns.set_palette("pastel")
ax = sns.countplot(y="Grade Level", hue="Gender", data=df_Science)
ax.legend(loc='upper right',frameon=True)
plt.title('Gender vs Grades', fontsize=18, fontweight='bold')
ax.set(xlabel='COUNT',ylabel='Grade Level')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
labels=['Not At Grade Level', 'Grade Level']
plt.pie(df_Science['Grade Level'].value_counts(),labels=labels,explode=[0.1,0.1],
        autopct='%1.2f%%',colors=['#E37383','#FFC0CB'], startangle=90)
plt.title('Grade Level')
plt.axis('equal')
plt.show()

print("Here we get to know the percentage wise how many percentage of students is at 'Garde Level' and how many students is not at Graded")
```

```python
plt.figure(figsize=(8,8))
plt.title('Correlation Analysis',color='Red',fontsize=20,pad=40)

corr = df_Science.corr()
mask = np.triu(np.ones_like(corr,dtype = bool))
sns.heatmap(df.corr(),mask=mask,annot=True,linewidths=.5);
plt.xticks(rotation=60)
plt.yticks(rotation = 60)
plt.show()
```

```python
sns.set_context("notebook",font_scale=0.8)
sns.kdeplot(data=df_Science, x="EL Score", hue="Gender", multiple="layer",fill=True)
plt.xlabel('EL Score')
plt.title('EL Score vs Gender Kde Plot',fontsize=15, fontweight='bold')

plt.show()
```
